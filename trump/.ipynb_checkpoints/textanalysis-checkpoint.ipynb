```python
from pymongo import MongoClient
from nltk import *

mongo = MongoClient('ds017678.mlab.com', 17678)
db = mongo['heroku_sjm9g8xc']
db.authenticate('heroku_sjm9g8xc', '4mvq9ej8mp5hlvqdqbgc53d2bf')

scraped = db.twitter
```

```python
import pandas as pd
import re
from nltk.corpus import stopwords

strings = [word_tokenize(re.sub('[^A-Za-z0-9 ]', '', i["Tweets"]).lower()) for i in scraped.find()]
avalues = [i['Favs'] for i in scraped.find()]
```

```python
stop_words = set(stopwords.words('english')) 

cleaned_strings = []

for i in strings:
    cleaned_strings.append([j for j in i if j not in stop_words])
    
cleaned_strings
```

```python
words = []
values = []
for i in range(len(strings)-1):
    for j in range(len(strings[i])):
        if strings[i][j] not in stop_words:
            words.append(strings[i][j])
            values.append(avalues[i])
        else:
            pass
twitter = pd.DataFrame(
    {
        "word":words
        , "value":values
    }
)
```

```python
import numpy as np

g =  twitter.groupby('word').agg([np.mean, 'count']).reset_index()

g.columns = ["word","mean","count"]

g.sort_values("mean", ascending=False)
```
