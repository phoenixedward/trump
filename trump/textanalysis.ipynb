```python
from pymongo import MongoClient
from nltk import *

mongo = MongoClient('ds017678.mlab.com', 17678)
db = mongo['heroku_sjm9g8xc']
db.authenticate('heroku_sjm9g8xc', '4mvq9ej8mp5hlvqdqbgc53d2bf')

scraped = db.twitter
```

```python
import pandas as pd
import re
from nltk.corpus import stopwords

strings = [word_tokenize(re.sub('[^A-Za-z0-9 ]', '', i["Tweets"]).lower()) for i in scraped.find()]
avalues = [i['Favs'] for i in scraped.find()]
```

```python
stop_words = set(stopwords.words('english')) 

cleaned_strings = []
cleaned_avalues = []


for i in range(len(strings)-1):
    k = [j for j in strings[i] if j not in stop_words]
    cleaned_strings.append(k)
    if k != '':
        cleaned_avalues.append(avalues[i])
        
corpus = [" ".join(word) for word in cleaned_strings]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(ngram_range=(1,2))

features = vectorizer.fit_transform(corpus)

pd.DataFrame(features.todense(),columns=vectorizer.get_feature_names())
```
